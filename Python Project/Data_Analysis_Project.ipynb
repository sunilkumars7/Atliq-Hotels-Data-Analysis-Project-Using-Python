```python
lower_limit_rev_gen, upper_limit_rev_gen
```

```python
df_bookings[(df_bookings['revenue_generated'] < lower_limit_rev_gen) | 
            (df_bookings['revenue_generated'] > upper_limit_rev_gen)]
```
